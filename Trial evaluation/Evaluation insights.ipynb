```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('trialeval.csv') #place data in root folder or change path
```

```python
data
```

```python
#guessing emotion eval:
s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q1'] == 'Strongly agree':
    s_agree += 1
  elif row['Q1'] == 'Agree':
    agree += 1
  elif row['Q1'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q1'] == 'Disagree':
    disag += 1
  elif row['Q1'] == 'Strongly disagree':
    s_disag += 1

guess_emo = [s_disag, disag, neither, agree, s_agree]

print(guess_emo)

norm_guess_emo = [i/sum(guess_emo) for i in guess_emo]

print(norm_guess_emo)
```

```python
#Kai empathy:

s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q2'] == 'Strongly agree':
    s_agree += 1
  elif row['Q2'] == 'Agree':
    agree += 1
  elif row['Q2'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q2'] == 'Disagree':
    disag += 1
  elif row['Q2'] == 'Strongly disagree':
    s_disag += 1

kai_emp = [s_disag, disag, neither, agree, s_agree]

print(kai_emp)

norm_kai_emp = [i/sum(kai_emp) for i in kai_emp]

print(norm_kai_emp)
```

```python
#Kai engagement:

s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q3'] == 'Strongly agree':
    s_agree += 1
  elif row['Q3'] == 'Agree':
    agree += 1
  elif row['Q3'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q3'] == 'Disagree':
    disag += 1
  elif row['Q3'] == 'Strongly disagree':
    s_disag += 1

kai_engag = [s_disag, disag, neither, agree, s_agree]

print(kai_engag)

norm_kai_engag = [i/sum(kai_engag) for i in kai_engag]

print(norm_kai_engag)
```

```python
#Others empathy:

s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q5'] == 'Strongly agree':
    s_agree += 1
  elif row['Q5'] == 'Agree':
    agree += 1
  elif row['Q5'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q5'] == 'Disagree':
    disag += 1
  elif row['Q5'] == 'Strongly disagree':
    s_disag += 1

others_emp = [s_disag, disag, neither, agree, s_agree]

print(others_emp)

norm_others_emp = [i/sum(others_emp) for i in others_emp]

print(norm_others_emp)
```

```python
#Others engagement:

s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q6'] == 'Strongly agree':
    s_agree += 1
  elif row['Q6'] == 'Agree':
    agree += 1
  elif row['Q6'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q6'] == 'Disagree':
    disag += 1
  elif row['Q6'] == 'Strongly disagree':
    s_disag += 1

others_engag = [s_disag, disag, neither, agree, s_agree]

print(others_engag)

norm_others_engag = [i/sum(others_engag) for i in others_engag]

print(norm_others_engag)
```

```python
#Usefulness:

s_disag=0
disag=0
neither=0
agree=0
s_agree=0

for i, row in data.iterrows():
  if row['Q6.1'] == 'Strongly agree':
    s_agree += 1
  elif row['Q6.1'] == 'Agree':
    agree += 1
  elif row['Q6.1'] == 'Neither agree nor disagree':
    neither += 1
  elif row['Q6.1'] == 'Disagree':
    disag += 1
  elif row['Q6.1'] == 'Strongly disagree':
    s_disag += 1

use = [s_disag, disag, neither, agree, s_agree]

print(use)

norm_use = [i/sum(use) for i in use]

print(norm_use)
```

```python
#personas chosen (apart from kai)

robert=0
gabrielle=0
arman=0
olivia=0


for i, row in data.iterrows():
  if 'Robert' in row['Q4']:
    robert += 1
  if 'Gabrielle' in row['Q4']:
    gabrielle += 1
  if 'Arman' in row['Q4']:
    arman += 1
  if 'Olivia' in row['Q4']:
    olivia += 1
personas = [robert, gabrielle, arman, olivia]

print(personas)

norm_personas = [i/sum(personas) for i in personas]

print(norm_personas)
```

###Comparison

guess emotion:

```python
labels = ['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly Agree']
guess_emo_df = pd.DataFrame(list(zip(labels, norm_guess_emo)),
               columns =['response', 'percentage'])
```

```python
#guess emotion
plt.rcParams.update({'font.size': 18})
ax = guess_emo_df.plot.bar(figsize=(15,8), title="Level of agreement/disagreement with the statement \n  'The chatbot was good at guessing my emotion'", 
                                             fontsize=18, rot=0, color='mediumblue', ylim=(0,0.55));

ticks = ['Strongly \n disagree', 'Disagree', 'Neither agree \n nor disagree', 'Agree', 'Strongly \n Agree']
ax.set_xticklabels(ticks);
ax.set_yticklabels(['0%', '10%', '20%', '30%', '40%', '50%']);
plt.legend(loc='upper left');
```

empathy:

```python
kai_emp_df = pd.DataFrame(list(zip(labels, norm_kai_emp)),
               columns =['response', 'percentage'])

others_emp_df = pd.DataFrame(list(zip(labels, norm_others_emp)),
               columns =['response', 'percentage'])

prev_emp_df = pd.DataFrame(list(zip(labels, norm_prev_emp)),
               columns =['response', 'percentage'])
```

```python
plt.rcParams.update({'font.size': 22})
fig, axes = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(25,8));
fig.suptitle("Level of agreement/disagreement with the statement \n 'The chatbot displayed empathy in its responses throughout the conversation'")
plt.setp(axes,  ylim=(0,0.8), xticklabels=ticks, yticklabels=['0%', '10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%'])
plt.rcParams.update({'font.size': 18})
prev_emp_df.plot.bar(ax=axes[0], color='orangered');#, title='Previous chatbot implementation');
axes[0].legend(loc='upper left');
axes[0].set_xticklabels(ticks, fontsize=18, rotation=45);
kai_emp_df.plot.bar(ax=axes[1], color='mediumblue');#, title='Our implementation: Kai persona');
axes[1].legend(loc='upper left');
axes[1].set_xticklabels(ticks, fontsize=18, rotation=45);
others_emp_df.plot.bar(ax=axes[2], color='mediumblue');#, title='Our implementation: other personas');
axes[2].legend(loc='upper left', fontsize=18);
axes[2].set_xticklabels(ticks, fontsize=18, rotation=45);
```
